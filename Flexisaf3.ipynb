```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load dataset
df = pd.read_csv('StudentsPerformance.csv')

# Check first few rows
print(df.head())
```

```python
# Check number of rows and columns
print(df.shape)
```

```python
#Check the Data Types
df.info()
```

```python
#Check for Missing Values
df.isnull().sum()
```

```python
# Summary Statistics for Numeric Columns
df.describe()
```

```python
# Unique Values in Categorical Columns
for col in df.select_dtypes(include='object').columns:
    print(f"{col} unique values:\n{df[col].unique()}\n")
```

```python
#  Correlation Matrix (for numeric variables)
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#  Create a Scatter Plot to visualize the relationship between Math and Reading scores
plt.scatter(df['math score'], df['reading score'])

# Add a title to the plot
plt.title('Math vs Reading Score')

# Label the X-axis as 'Math Score'
plt.xlabel('Math Score')

# Label the Y-axis as 'Reading Score'
plt.ylabel('Reading Score')

# Display the plot
plt.show()
```

```python
#  Create a Scatter Plot to visualize the relationship between Math and Reading scores
plt.scatter(df['math score'], df['reading score'])

# Add a title to the plot
plt.title('Math vs Reading Score')

# Label the X-axis as 'Math Score'
plt.xlabel('Math Score')

# Label the Y-axis as 'Reading Score'
plt.ylabel('Reading Score')

# Display the plot
plt.show()
```

```python
# Create a Histogram to visualize the distribution of Writing Scores
# 'kde=True' adds a smooth Kernel Density Estimate (curve) on top of the histogram
sns.histplot(df['writing score'], kde=True)

# Add a title to the plot
plt.title('Distribution of Writing Scores')

# Display the plot
plt.show()
```

```python
#  Create a Box Plot to visualize the distribution of Math scores grouped by Gender
sns.boxplot(x='gender', y='math score', data=df)

# Add a title to the plot
plt.title('Math Scores by Gender')

# Display the plot
plt.show()
```
